```python
#importing libraries
import requests
from bs4 import BeautifulSoup
import csv
```

```python
url='https://www.imdb.com/chart/top/'
```

```python
headers={'User-Agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'}
```

```python
response=requests.get(url,headers=headers)
```

```python
response
```

```python
soup=BeautifulSoup(response.content,'html.parser')
```

```python
soup.find_all('h3')[1].text
```

```python
h3=soup.find_all('h3')
```

```python
h3[1:10]
```

```python
for movie in h3:
    print(movie.text)
```

```python
movies=soup.find_all('li',class_='ipc-metadata-list-summary-item sc-59b6048d-0 cuaJSp cli-parent')
```

```python
for movie in movies:
    name=movie.find('div',class_='ipc-metadata-list-summary-item__c').h3.text.split('.')[1]
    rank=movie.find('div',class_='ipc-metadata-list-summary-item__c').h3.text.split('.')[0]
    year=movie.find('div',class_='sc-479faa3c-7 jXgjdT cli-title-metadata').span.text
    duration=movie.find('div',class_='sc-479faa3c-7 jXgjdT cli-title-metadata').find_all('span')[1].text
    rating=movie.find('div',class_='sc-e3e7b191-0 jlKVfJ sc-479faa3c-2 eUIqbq cli-ratings-container').span.text[0:4]
    print(rating)
    break
    
```

```python
with open('movie-data.csv','w',newline='',encoding='utf-8-sig') as f:
    csv_writer=csv.writer(f)
    heading =['Rank','Name','Year','Duration','Ratings']
    csv_writer.writerow(heading)
    for movie in movies:
        name=movie.find('div',class_='ipc-metadata-list-summary-item__c').h3.text.split('.')[1]
        rank=movie.find('div',class_='ipc-metadata-list-summary-item__c').h3.text.split('.')[0]
        year=movie.find('div',class_='sc-479faa3c-7 jXgjdT cli-title-metadata').span.text
        duration=movie.find('div',class_='sc-479faa3c-7 jXgjdT cli-title-metadata').find_all('span')[1].text
        rating=movie.find('div',class_='sc-e3e7b191-0 jlKVfJ sc-479faa3c-2 eUIqbq cli-ratings-container').span.text[0:4]
        csv_writer.writerow([rank,name,year,duration,rating])
```
